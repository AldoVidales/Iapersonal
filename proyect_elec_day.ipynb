```python
datos['school_day'] = datos['school_day'].map({'N':0, 'Y': 1})
datos['holiday'] = datos['holiday'].map({'N':0, 'Y': 1})
```

```python
datos['date']=pd.to_datetime(datos['date'])
```

```python
datos.head()
```

```python
datos
```

```python
datos['date']=np.arange(len(datos['date']))


```

```python
datos.info()
```

```python
from sklearn.neighbors import  KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge
x=datos[['RRP','RRP_positive','RRP_negative','min_temperature','max_temperature','solar_exposure','rainfall','school_day','holiday','frac_at_neg_RRP']]
y=datos[['demand','demand_pos_RRP','demand_neg_RRP']]
x_ent,x_test,y_ent,y_test=train_test_split(x,y,test_size=0.002)
knn=KNeighborsRegressor(n_neighbors=3)
knn.fit(x_ent,y_ent)


datos[['predict_d','predict_dpr','predict_dnr']]=knn.predict(x)

knn.score(x_test,y_test)
```

Real vs predict

```python
plt.scatter(datos['date'],datos['predict_d'],label='demandpredict')
plt.scatter(datos['date'],datos[['demand']],label='demand')
plt.legend()
plt.show()
```

```python
plt.scatter(datos['date'],datos[['predict_dpr']],label='predictdemandpr')
plt.scatter(datos['date'],datos[['demand_pos_RRP']],label='demandpr')

plt.legend()
plt.show()
```

```python
plt.scatter(datos['date'],datos[['predict_dnr']],label='predictdemandnr')
plt.scatter(datos['date'],datos[['demand_neg_RRP']],label='demandnr')
plt.legend()
plt.show()
```
